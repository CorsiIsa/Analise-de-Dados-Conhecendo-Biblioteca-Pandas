# Imóveis residenciais

```python
import pandas as pd
```

```python
url = 'data/aluguel.csv'

dados = pd.read_csv(url, sep=';')

dados
```

Primeiro iremos criar uma lista com os tipos de imóveis sem repetições.

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residenciais = ['Apartamento', 'Casa', 'Casa de Condomínio', 'Casa de Vila', 'Quitinete']
```

Agora se passarmos essa lista no método **isin()**, ele irá marcar como True todas as vezes que algum dos nossos elementos aparecerem no nosso DataFrame.

```python
selecao = dados['Tipo'].isin(residenciais)
```

Agora iremos passar esse seleção para dentro do nosso DataFrame dados, onde ele irá apagar todos os registros False e deixará apenas os registros com True.

```python
dados_residencial = dados[selecao]

dados_residencial
```

## Exportando DataFrame

Após essa filtragem, precisamos exportar esse novo DataFrame que criamos. Com o método **to_csv**

```python
dados_residencial.to_csv('data/aluguel_residencial.csv', sep=';', index= False)
```

## Seleções e Frequências de Imóveis

Agora iremos selecionar os registros classificados como 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.
