# Criando Agrupamentos

```python
import pandas as pd
```

```python
url = 'data/aluguel_residencial.csv'

dados = pd.read_csv(url, sep=';')

dados
```

Vamos ver o valor médio da coluna Valor.

```python
dados['Valor'].mean()
```

Agora iremos analisar qual o valor médio de acordo com o bairro do registro.

Vamos primeiro criar uma lista com alguns bairros que iremos analisar.

```python
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']

selecao = dados['Bairro'].isin(bairros)

dados = dados[selecao]

dados
```

Aqui criaremos um grupo de acordo com os bairros que selecionamos, ou seja utilizaremos a função **groupby()** que irá pegar todos os índices e juntar em um dicionário.

```python
grupo_bairro = dados.groupby('Bairro')

grupo_bairro.groups
```

Agora iremos buscar a nossa média de valor por bairro, da seguinte forma:

```python
for bairro, data in grupo_bairro:
    print('{} -> {}'.format(bairro, data.Valor.mean()))
```

Podemos fazer da seguinte forma também:

```python
grupo_bairro[['Valor', 'Condominio']].mean().round(2)
```

# Estatísticas Descritivas

Usamos a função **describe()** no grupo que criamos, conseguimos visualizar um relatório estatístico com informações básicas.

```python
grupo_bairro['Valor'].describe().round(2)
```

Tem como filtrar essas estatísticas e pegar apenas as que nos interessam, utilizando a função **aggregate()**.

```python
grupo_bairro['Valor'].aggregate(['min', 'max', 'sum'])
```

Se não tivessemos selecionado o tipo de bairro que queriamos, iria ficar muita informação dificultando a visualização dos dados, sendo assim nessas situações podemos utiliza a biblioteca **Matplotlib**.

```python
import matplotlib.pyplot as plt
```

```python
url = 'data/aluguel_residencial.csv'

dados = pd.read_csv(url, sep=';')

dados
```

```python
grupo_bairro = dados.groupby('Bairro')

grupo_bairro.groups
```

Agora iremos criar um gráfico para visualizarmos melhor as informações.

```python
plt.rc('figure', figsize = (20,10))
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color='blue')
```

Agora vamos configurar esse gráfico para ficar melhor de entender as informações.

Vamos dar um nome para o valor Y, utilizando a função **ylabel()**.

```python
fig.set_ylabel('Valor do Aluguel')
```

Agora vamos dar um título, utilizando a função **set_title()**.

```python
fig.set_title('Valor médio do aluguel por bairro', {'fontsize': 22})
```
